# **‚Ä¢	Data Preparation dengan One Hot Encoding**

```python
df_lasttable = df.iloc[:,-1]
df_lasttable = pd.DataFrame(df_lasttable)
df_lasttable.head()

encoder = OneHotEncoder()

encoded_df = pd.DataFrame(
    encoder.fit_transform(df_lasttable).toarray(),
    columns=encoder.get_feature_names_out(df_lasttable.columns)
)

df_encoded = pd.concat([df['target'], encoded_df], axis=1)
danish05.info()
df_encoded
```

# **‚Ä¢	Data Preparation dengan Outlier Removal**

```python
z_score = np.abs((df['max heart rate'] - df['max heart rate'].mean()) / df['max heart rate'].std())
threshold = 3
outliers = df.loc[round(z_score) > threshold]
df_cleaned = df.loc[round(z_score) <= threshold]

df_encodez = df.copy()

df_encodez = df_encodez.drop(columns=['max heart rate'])
df_encodez['max heart rate'] = z_score

print(z_score)
danish05.info()
df_encodez
```
