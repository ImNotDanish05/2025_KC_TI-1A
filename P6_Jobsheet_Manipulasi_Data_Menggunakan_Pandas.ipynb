## Soal-1
Buatlah table list Top 5 Customers yang memiliki nilai transaksi paling banyak
---





```python
# 1. Hitung berapa banyak
top_5_customer_ids = df_transactions_1k['CustomerID'].value_counts().head(5)

# 2. Filter df_merge berdasarkan CustomerID yang ada di top 5
df_merged2 = pd.merge(top_5_customer_ids, df_merged, on='CustomerID', how='left')

# 3. Menghapus baris duplikat
df_merged2 = df_merged2.drop_duplicates(subset='CustomerID')

# 4. Tampilkan data hasil filter
df_merged2
```

## Soal-2
Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak
---

```python
# 1. Urutkan top 5 berdasarkan GasStationID
top_5_gasstation_ids = df_transactions_1k['GasStationID'].value_counts().head(5)

"""
2. Di merge antara top_5_gasstation_ids dengan df_merged berdasarkan GasStationID
yang diprioritaskan top_5_gasstation_ids (how=left) dan disimpan di df_merged3
"""
df_merged3 = pd.merge(top_5_gasstation_ids, df_merged, on='GasStationID', how='left')

# 3. Hitung total Amount per GasStationID
amount_sum = df_merged.groupby('GasStationID')['Amount'].sum().reset_index()
amount_sum.rename(columns={'Amount': 'Amount'}, inplace=True)

# 4. Merge dengan df_merged3 berdasarkan GasStationID
df_merged3 = pd.merge(df_merged3.drop(columns='Amount', errors='ignore'), amount_sum, on='GasStationID', how='left')

# 5. Drop duplikat
df_merged3 = df_merged3.drop_duplicates(subset='GasStationID')

# 6. Print HANYA Description, Amount, dan Price
df_merged3[['Description', 'Amount', 'Price']]
```

## Soal-3
Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak
---

```python
# 1. Menjumlahkan berapa banyak baris per ProductID dan diurutkan top 5
top_5_product_ids = df_transactions_1k['ProductID'].value_counts().head(5)

# 2. Mergekan df_merged dengan top_5_product_ids di df_merged5
df_merged4 = pd.merge(top_5_product_ids, df_merged, on='ProductID', how='left')

# 3. Rename
df_merged4.rename(columns={'Description': 'Products'}, inplace=True)
df_merged4.rename(columns={'count': 'Transiction Count'}, inplace=True)

# 4. Jumlah Amount per setiap ProductID. dan reset_index() jangan lupa supaya file tetap bisa dibaca sebagain DataFrame
amount_gas = df_merged.groupby('ProductID')['Amount'].sum().reset_index()
# 5. Lalu merge denga drop Amount supaya direplace dengan Amount dari amount_gas (yang sum())
df_merged4 = pd.merge(df_merged4.drop(columns='Amount', errors='ignore'), amount_gas, on='ProductID', how='left')

# 6. Hitung rata-rata harga per ProductID
avg_price_per_product = df_merged.groupby('ProductID')['Price'].mean().reset_index()
avg_price_per_product.rename(columns={'Price': 'AvgPrice'}, inplace=True)

# 7. Merge ke df_merged4
df_merged4 = pd.merge(df_merged4, avg_price_per_product, on='ProductID', how='left')

# 8. Hitung Sales
df_merged4['Sales'] = df_merged4['Amount'] * df_merged4['AvgPrice']
df_merged4 = df_merged4.round(0)

# 9. Drop duplicate
df_merged4 = df_merged4.drop_duplicates(subset='ProductID')

# 10. Print dengan selection
df_merged4[['Products', 'Transiction Count', 'Amount', 'Sales']]
```

## Soal-4
Buatlah deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset transaction_1k
---

```python
# 1. Format tanggal pastikan benar
df_merged5 = df_merged.copy()
df_merged5['Date'] = pd.to_datetime(df_merged5['Date'])

# 2. Data hanya untuk tanggal 23 hingga 26
df_merged5 = df_merged5[
    (df_merged5['Date'] >= '2012-08-23') &
     (df_merged5['Date'] <= '2012-08-26')
     ]

# 3. Karena kebetulan tanggal hanya d3 hingga 26, maka saya langsung
result = df_merged5.groupby('Date').agg(
    Amount_count=('Amount', 'count'),
    Amount_mean=('Amount', 'mean'),
    Amount_std=('Amount', 'std'),
    Amount_min=('Amount', 'min'),
    Amount_25=('Amount', lambda x: x.quantile(0.25)),
    Amount_50=('Amount', 'median'),
    Amount_75=('Amount', lambda x: x.quantile(0.75)),
    Amount_max=('Amount', 'max'),
    Sales_count=('Price', 'count'),
    Sales_mean=('Price', 'mean'),
    Sales_std=('Price', 'std'),
    Sales_min=('Price', 'min'),
    Sales_25=('Price', lambda x: x.quantile(0.25)),
    Sales_50=('Price', 'median'),
    Sales_75=('Price', lambda x: x.quantile(0.75)),
    Sales_max=('Price', 'max')
)

# 4. Buat tabel jadi lebih rapi dengan ada main column seperti amount dan sales
result.columns = pd.MultiIndex.from_tuples([
    ('Amount', 'count'),
    ('Amount', 'mean'),
    ('Amount', 'std'),
    ('Amount', 'min'),
    ('Amount', '25%'),
    ('Amount', '50%'),
    ('Amount', '75%'),
    ('Amount', 'max'),
    ('Sales', 'count'),
    ('Sales', 'mean'),
    ('Sales', 'std'),
    ('Sales', 'min'),
    ('Sales', '25%'),
    ('Sales', 'median'),
    ('Sales', '75%'),
    ('Sales', 'max')
])

# 5. Print dong :D
result = result.reset_index()
result
```

## Soal-5
Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi
---

```python
# 1. Backup file seperti biasa :V
df_merged6 = df_merged.copy()

# 2. Identifikasikan Date dan Time, untuk memastikan mereka sudah pada format yang kita mau
df_merged6['Date'] = pd.to_datetime(df_merged['Date'])
df_merged6['Time'] = pd.to_datetime(df_merged['Time'], format='%H:%M:%S').dt.hour

# 3. Lalu kita groupby untuk setiap jam di satu hari akan di total berapa banyak Amount dan Price
df_grouped = df_merged6.groupby([df_merged6['Date'].dt.date, 'Time']).agg(
    Total_Sales=('Amount', 'sum'),
    Gas_Sales=('Price', 'sum')
).reset_index()

# 4. Rename & Print
df_grouped2 = df_grouped.rename(columns={'Date': 'Date', 'Time': 'Time', 'Total_Sales': 'Amount', 'Gas_Sales': 'Gas Sales'})
df_grouped2
```

```python
# 1. Hitung ada berapa banyak transaksi pada setiap jam pada 1 hari.
# Misal jam 9 ada 3 transaksi, maka 3 itu akan jadi value dari User_Count pada Time 9
df_user_count = df_merged6.groupby([df_merged6['Date'].dt.date, 'Time']).size().reset_index(name="User_Count")

# 2. Mengurutkan Date dari kecil ke besar (True), dan User_Count sebaliknya (False)
# drop duplicate 0, maka dari itu User_Count hanya mengeprint baris yang paling atas. Otomatis mengeprint tabel yang User_Countnya besar
df_peak_hour = df_user_count.sort_values(['Date', 'User_Count'], ascending=[True, False]).drop_duplicates('Date')

# 3. Menggroupby untuk mentotal Amount dan Gas Sales seperti biasa
df_grouped = df_merged6.groupby([df_merged6['Date'].dt.date, 'Time']).agg(
    Total_Sales=('Amount', 'sum'),
    Gas_Sales =('Price', 'sum')
).reset_index()

# 4. Di merge dengan df_peak_hour
df_result = pd.merge(df_peak_hour, df_grouped, on=['Date', 'Time'], how='left')

# 5. Rename & Printout
df_result = df_result.rename(columns={'Date': 'Date', 'Time': 'Time', 'User_Count': 'User Count', 'Total_Sales': 'Amount', 'Gas_Sales': 'Gas Sales'})
df_result
```
