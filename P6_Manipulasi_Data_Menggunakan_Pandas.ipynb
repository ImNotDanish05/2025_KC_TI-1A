## Mengatasi Missing Values dengan Drop Data

### 1. Hapus Data yang kosong (NaN)

```python
df_bersih = df.dropna()
df_bersih
```

## Mengatasi Missing Values dengan Nilai Baru

### 2. Ganti Data yang kosong (NaN) menjadi :

### <ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
df2 = df.copy()
df2['Gaji'] = df2['Gaji'].fillna(25000000)
df2['Umur'] = df2['Umur'].fillna(50)
df2
```

## Mengatasi Missing Values dengan Nilai Mean

## 3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df3 = df.copy()
df3['Gaji'] = df3['Gaji'].fillna(df3['Gaji'].mean())
df3['Umur'] = df3['Umur'].fillna(df3['Umur'].mean())
df3
```

```python
df31 = df3.copy()
df31['Umur'] = df3['Umur'].round(0).astype(int)
df31['Gaji'] = df3['Gaji'].round(0).astype(int)
df31
```

## Mengatasi Missing Values dengan Nilai Median

## 4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df4 = df.copy()
df4['Gaji'] = df4['Gaji'].fillna(df4['Gaji'].median()).round(0).astype(int)
df4['Umur'] = df4['Umur'].fillna(df4['Umur'].median()).round(0).astype(int)
df4
```

## Mengatasi Missing Values dengan Nilai Mode

## 5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df5 = df.copy()
df5['Gaji'] = df5['Gaji'].fillna(df5['Gaji'].mode()).round(0).astype(int)
df5['Umur'] = df5['Umur'].fillna(df5['Umur'].mode()).round(0).astype(int)
df5
```

## Mengatasi Wrong Data dengan Ubah Data

## 6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
df6 = df.copy()
df6.loc[df6['Umur'] > 120, 'Umur'] = 33
df6
```
