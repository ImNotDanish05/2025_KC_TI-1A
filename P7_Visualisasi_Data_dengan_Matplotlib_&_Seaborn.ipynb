```python
# 1.	Melakukan Visualisasi Data menggunakan Pie Chart
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Strawberry', 'Mint', 'Cookies and Cream')
votes = (2200, 1800, 1500, 1200, 1000)

colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D','#D53032')

plt.pie(votes, labels=flavor)

explode = (0, 0, 0, 0.2, 0)

plt.pie(votes,
        labels = flavors,
        autopct = '%1.2f%%',
        colors = colors,
        explode = explode,
        shadow = True)
plt.show()
```

```python
#2.	Melakukan Visualisasi Data menggunakan Bar Chart
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

countries = ('Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela')
population = (45076704, 11382894, 211995908, 19087969, 50568480, 17448392, 782775, 71089, 32880332, 584024, 3470413, 28889459)

# First Route
# Ini bikin dataframe pakai pandas. Dataframe itu seperti tabel
df = pd.DataFrame({'countries': countries, 'population': population})
"""
👉 Ini yang bikin data jadi urut dari populasi terbesar ke terkecil!
by='population': berarti urut berdasarkan kolom population.
ascending=False: berarti dari besar ke kecil.
inplace=True: supaya dataframe-nya langsung diubah (nggak bikin salinan baru).
"""
df.sort_values(by='population', ascending=False, inplace=True)
#👉 Ini bikin urutan angka dari 0 sampai jumlah negara - 1. Misalnya ada 12 negara, maka x_coords = [0 1 2 3 ... 11]. Ini dipakai buat sumbu X di bar chart.
x_coords = np.arange(len(countries))
#👉 Ini bikin bar chart (diagram batang).
plt.bar(x_coords, df['population'], tick_label=df['countries'])
"""
👉 Ini bagian dekorasi:
xticks(rotation=90): biar nama negaranya ditulis vertikal (biar muat).
ylabel: ngasih label ke sumbu Y.
title: kasih judul grafik.
show(): tampilkan grafiknya.
"""
plt.xticks(rotation=90) #rotates text for x-axis labels
plt.ylabel('Population')
plt.title('South American Populations')
plt.show()

# Second Route
plt.bar(countries, population)
plt.xticks(rotation=90)
plt.xlabel('Countries')
```
